# FUNCIONES 

```python
def miFuncion():
    print("Hola")
miFuncion()
```

```python
def saludoPers(nombre):
    print("Hola", nombre)
saludoPers("Simon")
```

```python
saludoPers("Maria")
saludoPers("jose")
```

```python
def SaludoConFunc(funcion, nombre):
    funcion(nombre)
SaludoConFunc(saludoPers, "Mario")
    
```

```python
def elevarpoten(numero, potencia):
    return numero ** potencia
elevarpoten(2,3)
```

```python
def MiFuncDeten(numero):
    if numero is None:
        return
    else:
        print(numero)
numero = 11
MiFuncDeten(numero)
```

```python
def isprimo(numero):
    if numero / 1 == numero and numero / numero == 1:
        print("El número es primo")
    else:
        print("el numero no es primo")
numero = 10
isprimo(numero)
```
