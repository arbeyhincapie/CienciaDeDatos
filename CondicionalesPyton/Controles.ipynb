```python
def miFuncion():
    print("Hola")
miFuncion()
```

```python
def saludoPers(nombre):
    print("Hola", nombre)
saludoPers("Simon")
```

```python
saludoPers("Maria")
saludoPers("jose")
```

```python
def SaludoConFunc(funcion, nombre):
    funcion(nombre)
SaludoConFunc(saludoPers, "Mario")
    
```

```python
def elevarpoten(numero, potencia):
    return numero ** potencia
elevarpoten(2,3)
```

```python
def MiFuncDeten(numero):
    if numero is None:
        return
    else:
        print(numero)
numero = 11
MiFuncDeten(numero)
```

```python
def isprimo(numero):
    if numero / 1 == numero and numero / numero == 1:
        print("El número es primo")
    else:
        print("el numero no es primo")
numero = 10
isprimo(numero)
```

# Funcion con parametros de lista

```python
def FuncionLista(lista):
    for elemento in lista:
        print(elemento)
        
milista = [1,2,3,"donald", True]
FuncionLista(milista)
```

```python
def funcionparadef(numero, potencia = 2):
    return numero ** potencia
funcionparadef(2)
```

```python
funcionparadef(2,3)
```

```python
def promedio(valores):
    suma = 0
    for elemento in valores:
        suma += elemento
    division = suma / len(valores)
    return division

valores = [2,4,6,10]
print(promedio(valores))
    
    
```

```python
def mediana(valores):
    sorted(valores)
    if len(valores) % 2 ==0:
        n=len(valores)
        mediana = ((valores[n/2-1]) + (valores[n/2]))/2
    else:
        mediana = valores[len(valores)/2]
        

        
valores = [2,4,5,6,7,7,9,9,10,11]  
print(mediana(valores))

        
```
